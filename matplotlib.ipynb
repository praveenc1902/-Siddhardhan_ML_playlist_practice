```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
z=  np.cos(x)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin wave')
plt.show()
#we didnt mention which type of plot we need so, it has given line plot
```

```python
x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y,'*','red')
plt.show()
```

#Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
languages = ['English','French','Latin','Spanish']
people = [100,200,12,44]
ax.bar(languages,people)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
languages = ['English','French','Latin','Spanish']
people = [100,200,12,44]
ax.pie(people,labels=languages,autopct='%1.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,20)
y = np.sin(x)
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.scatter(x,y,marker='*')
plt.show()
```
