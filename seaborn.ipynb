```python
from sklearn.datasets import load_boston
```

```python
bs = load_boston()
dir(bs)
```

```python
df = pd.DataFrame(bs.data, columns = bs.feature_names)
df['PRICE'] =  bs.target
df.head()
```

```python
sns.distplot(df['PRICE'])
```

```python
corr = df.corr()
print(corr)
```

```python
sns.heatmap(corr,annot=True,annot_kws={'size':9})
```
